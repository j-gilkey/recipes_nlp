```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, ward
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
import pre_processing
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import pyLDAvis
import pyLDAvis.sklearn
```

```python
df = pd.read_csv('concat_uncleaned_recipes.csv').dropna()
df = pre_processing.filter_cuisines(df, [2, 3, 5, 7, 8, 9, 13])
df['Ingredients'] = df.apply(lambda row: ' '.join(pre_processing.clean_strings(row['Ingredients'])), axis=1)

#text_data = list(df['Ingredients'])

lda_tf = LatentDirichletAllocation(n_components=8, random_state=0)

secondary_stop_words = ['ingredient', 'list', 'add', 'finely', 'available']

vec = CountVectorizer(stop_words=secondary_stop_words)
X = vec.fit_transform(df['Ingredients'])

lda_tf.fit(X)
pyLDAvis.enable_notebook()
pyLDAvis.sklearn.prepare(lda_tf, X, vec)
```

```python
df = pd.read_csv('concat_uncleaned_recipes.csv').dropna()
df = pre_processing.filter_cuisines(df, [2, 3, 5, 7, 8, 9, 13])
df['Ingredients'] = df.apply(lambda row: ' '.join(pre_processing.clean_strings(row['Ingredients'])), axis=1)

#text_data = list(df['Ingredients'])

lda_tf = LatentDirichletAllocation(n_components=7, random_state=0)

secondary_stop_words = ['ingredient', 'list', 'add', 'finely', 'available']

vec = CountVectorizer(stop_words=secondary_stop_words)
X = vec.fit_transform(df['Ingredients'])

lda_tf.fit(X)

lda_output = lda_tf.transform(X)

topicnames = ['Topic_' + str(i) for i in range(lda_tf.n_components)]

df_document_topic = pd.DataFrame(np.round(lda_output, 2), columns=topicnames, index=df.index)

dominant_topic = np.argmax(df_document_topic.values, axis=1)
df_document_topic['dominant_topic'] = dominant_topic

print(df_document_topic.head)
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.sklearn.prepare(lda_tf, X, vec)
```
